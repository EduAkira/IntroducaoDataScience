#Analisando as notas em geral

```python
import pandas as pd #importando blioteca pandas para leitura de arquivo em formato csv (execel)
#documentação -> pandas.DataFrame

import seaborn as sns
url = ("https://raw.githubusercontent.com/EduAkira/IntroducaoDataScience/main/Codigo/Introdução_a_Data_Science.ipynb")


notas = pd.read_csv("ratings.csv") #atribuinto a notas o caminho e qual a arquivo csv a ler
notas.head() #head motra as primeiras nodas (cabesçalho)
```

```python
notas.shape #formato da tabela (avilações/colunas)
```

```python
notas.columns = ["usuarioId", "filmesId", "nota", "momento"]  #troca da nomenclatuta das colunas
notas.head()
```

```python
notas["nota"].unique() #mostra apenas uma coluna especifica/ .unique() mostra quais os valores que tem na coluna, nao a quantidade
```

```python
notas["nota"].value_counts() #conta a quantidade dos elementos
```

```python
print("media", notas["nota"].mean())
print("mediana", notas.nota.median())
```

```python
notas.nota.head()
```

```python
notas.nota.plot(kind="hist")  #impressao na tela de forma mais visual/ grafica  conhecido como plot
```

```python
notas.nota.describe() #descrição dos dados
```

```python
sns.boxplot(notas.nota)
```

#Analisando notas  de forma especificas por filme

```python
notas.query("filmesId == 1").nota.mean()  #== significa igual / query e como se fosse uma pergunta no parametro
```

```python
media_por_filme = notas.groupby("filmesId").mean()["nota"] #agrupamento de informações no data frame para melhor visualização, retorna um dataFrame. dumentação DataFrameGroupBy
media_por_filme.head()
```

```python
media_por_filme.plot(kind="hist")
```

```python
sns.boxplot(y=media_por_filme)
```

```python
sns.histplot(media_por_filme, kde=True)
```

```python
media_por_filme.describe()
```

```python
import matplotlib.pyplot as plt

plt.hist(media_por_filme)
plt.title("histograma das médias dos filmes")
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=media_por_filme)
```

#Estudo de variaveis

```python
tmdb = pd.read_csv("tmdb_5000_movies.csv")
tmdb.head()
```

```python
tmdb.original_language.unique()  #variavei sem ordem de hierarquia podem ser chamadas de categoricas, no caso de lingua sao categoricas nominal
```

```python
tmdb.original_language.value_counts().index
```

```python
tmdb.original_language.value_counts().values
```

```python
contagem_de_lingua = tmdb.original_language.value_counts().to_frame().reset_index()
contagem_de_lingua.columns = ["original_language", "total"]
contagem_de_lingua.head()
```

```python
sns.barplot(x = "original_language", y = "total", data = contagem_de_lingua)
```

```python
sns.catplot(x = "original_language", kind="count", data = tmdb)
```
